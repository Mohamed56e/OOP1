```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


class Student(Person):
    def __init__(self, name, age, student_id, gpa, major):
        super().__init__(name, age)
        self.student_id = student_id
        self.gpa = gpa
        self.major = major
        self.grades = {}

    def add_grade(self, subject, grade):
        self.grades[subject] = grade
    def calculate_gpa(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades.values()) / len(self.grades)   
    def print_all(self):
        print("name = ", self.name)
        print("age = " ,self.age)
        print("student_id = ", self.student_id)
        print("gpa = ",self.calculate_gpa())
        print("major = ", self.major)
        print("grades = ", self.grades)
class Teacher(Person):
    def __init__(self, name, age, teacher_id, department, salary):
        super().__init__(name, age)
        self.teacher_id = teacher_id
        self.department = department
        self.salary = salary

    def total_salary(self, target):
        return self.salary + target
   
class Course:
    def __init__(self, name_subject, id_subject, teacher):
        self.name_subject = name_subject
        self.id_subject = id_subject
        self.teacher = teacher
        self.list_student = []
       
    
    def student_add(self, name_student):
        self.list_student.append(name_student)
    def student_delet(self, name_student):
        self.list_student.remove(name_student)
    
    
    def print_all(self):
        print("name_subject = ", self.name_subject)
        print(" id_subject = " ,self. id_subject)
        print("teacher = ", self.teacher.name)
        print("students:")
        if len(self.list_student) == 0:

            print("No Students")

        else:

            for student in self.list_student:

                print("-", student.name)    
```

```python
# إنشاء طالبين
s1 = Student("Mohamed", 20, 1, 3, "AI")
s2 = Student("Ahmed", 21, 2, 3.5, "Cyber Security")

# إضافة درجات
s1.add_grade("Python", 90)
s1.add_grade("Math", 80)

s2.add_grade("Python", 95)
s2.add_grade("Math", 85)

# إنشاء مدرس
t1 = Teacher("Dr.Ahmed", 45, 101, "Artificial Intelligence", 20000)

# إنشاء كورس
c1 = Course("Python", 501, t1)

# إضافة الطلاب للكورس
c1.student_add(s1)
c1.student_add(s2)

# طباعة بيانات الطالب الأول
print("===== Student 1 =====")
s1.print_all()

print()

# طباعة بيانات الطالب الثاني
print("===== Student 2 =====")
s2.print_all()

print()

# طباعة بيانات الكورس
print("===== Course =====")
c1.print_all()
```
