```python

class Employee:
    Employee_num = 0
    def __init__(self, name, age, gender, salary):
        self.name = name 
        self.age = age
        self.gender = gender
        self.salary = salary
        Employee.Employee_num +=1
        
    def totalsalary(self,target):
        return self.salary + target
        
    def printall(self):
        print("name = ",self.name)
        print("age = ",self.age)
        print("gender = ",self.gender)
        print("salary = ",self.salary)
    
em1 = Employee("ahmed",22,"male",20000,)
em1.printall()
print("totalsalary = ", em1.totalsalary(1000))


em2 = Employee("mohamed",22,"male",25000,)
em2.printall()
print("totalsalary = ", em2.totalsalary(1000))


print("Employee.Employee_num = ",Employee.Employee_num)
```

```python
class BankAccount:
    def __init__(self, onwer, balance):
        self.onwer = onwer
        self.balance = balance

    def deposit(self, target):
        return self.balance + target
    def witgdraw(self, ammount):   
        if ammount <=  self.balance:
            self.balance -= ammount
            return self.balance
        else:
            print("Insufficient balance")
    def printall(self):
        print("Owner:", self.onwer)
        print("Balance:", self.balance)

acc1 = BankAccount("mohamed",20000)
acc1.deposit(500)
acc1.witgdraw(1000)
acc1.printall()


acc1 = BankAccount("",20000)
acc1.deposit(500)
acc1.witgdraw(1000)
acc1.printall()
```
