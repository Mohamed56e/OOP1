```python
from abc import ABCMeta, abstractmethod
class Animal(metaclass=ABCMeta):
    @abstractmethod
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed
     @abstractmethod   
    def getsound(self):
        pass
class Dog(Animal):
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed
     
    def getsound(self):
        pass

class Cat(Animal):
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed 
    def getsound(self):
        pass
dog1 = Dog("rot", 50)
print(dog1.name)
```
